###15.What is causation? Explain difference between correlation and causation with an example.

      ANS:

       If A causes B, it means that when A happens,

       doesn’t necessarily cause the other.

###16.What is an Optimizer? What are different types of optimizers? Explain each with an example.
       ANS:

       An optimizer is a tool in machine learning
       that helps the model improve by adjusting
       its parameters (like weights) to minimize errors.


      Types of Optimizers:

        1.Gradient Descent:
       Basic optimizer that adjusts parameters
       based on the error’s slope.

       Example: If you're climbing a hill, gradient
       descent helps you go down the steepest slope
       to reach the bottom (minimum error).

       2.Stochastic Gradient Descent (SGD):

       A faster version of gradient descent. It uses one
       random data point to update the model,
       making it quicker but a bit more unstable.
       Example: Like taking random small steps to
       find the bottom of a hill.

       3.Momentum:

       Adds "momentum" to the updates, so the model
       keeps moving in the same direction even when
       the path is unclear.

      Example: It’s like rolling a ball downhill—it doesn’t
      stop easily, helping it go faster.

      4.AdaGrad:

      Adjusts the learning rate for each feature,
      giving smaller updates for common features
      and bigger updates for rare ones.

      Example: If a feature shows up often, it gets
      smaller updates. Rare features get bigger updates.

      5.RMSProp:

      Similar to AdaGrad but with a trick to avoid
      too much slowing down of learning over time.
      Example: Like keeping recent changes in mind
      while making updates, so the model keeps learning efficiently.

     6.Adam:
     Combines the benefits of Momentum and RMSProp.
      It adapts the learning rate based on past gradients
      and is very popular.

     Example: It’s like having a smart optimizer that
     knows when to adjust the speed and direction based
     on your past moves.

###25.explain data encoding?
       ans:
        Data encoding is the process of converting
        non-numeric data  or  into numbers.

        1.Label Encoding:
         Converts each unique category into a number.

        Example: "Red" becomes 0, "Green" becomes 1,
         "Blue" becomes 2.

        2.One-Hot Encoding:
        Creates a new column for each unique
        category and uses 1 or 0 to represent
        if the category is present.
      
        Example: For "Red," it becomes [1, 0, 0],
         for "Green," it becomes [0, 1, 0], and for
         "Blue," [0, 0, 1].

        3.Ordinal Encoding:
        Ordinal Encoding gives a number based on the
        order of categories.
       
        4.Target Guided Ordinal Encoding:
        Target Guided Ordinal Encoding gives a number
        based on how the categories relate to the target.
